```python
# Feature Engineering
```

```python
df['Year'] = df['ORDERDATE'].dt.year
df['Month'] = df['ORDERDATE'].dt.month
df['Day'] = df['ORDERDATE'].dt.day
df['Weekday'] = df['ORDERDATE'].dt.day_name()
```

```python
# Total Profit
df['Total_Profit'] = df['SALES'] - (df['QUANTITYORDERED'] * df['MSRP'])
```

```python
# Encode Categorical Variables
from sklearn.preprocessing import LabelEncoder
label_cols = ['STATUS', 'PRODUCTLINE', 'COUNTRY', 'DEALSIZE', 'Weekday']

for col in label_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
```

```python
# Scaling 
from sklearn.preprocessing import StandardScaler

num_cols = ['QUANTITYORDERED', 'PRICEEACH', 'SALES','DAYS_SINCE_LASTORDER', 'MSRP', 'Total_Profit']
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
# Target variable
y_sales = df['SALES']

# Use only these 6 columns to match Streamlit input
features_for_model = ['ORDERNUMBER', 'QUANTITYORDERED', 'PRICEEACH',
                      'ORDERLINENUMBER', 'DAYS_SINCE_LASTORDER', 'MSRP']

X_sales = df[features_for_model]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_sales, y_sales, test_size=0.2, random_state=42
)
```

```python
from sklearn.ensemble import RandomForestRegressor

model_sales = RandomForestRegressor(n_estimators=100, random_state=42)
model_sales.fit(X_train, y_train)

model_sales.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

y_pred = model_sales.predict(X_test)

# Metrics
print(f"MAE:  {mean_absolute_error(y_test, y_pred):.2f}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_test, y_pred)):.2f}")
print(f"R²:   {r2_score(y_test, y_pred):.2f}")
```

```python
# Hyper parameter tuning
param_grid = {
    'n_estimators': [100, 200, 300],          # number of trees
    'max_depth': [None, 10, 20, 30],          # tree max depth
    'min_samples_split': [2, 5, 10],           # min samples to split a node
    'min_samples_leaf': [1, 2, 4],             # min samples per leaf
    'max_features': ['auto', 'sqrt']            # features considered at each split
}
```

```python
from sklearn.model_selection import GridSearchCV

rf = RandomForestRegressor(random_state=42)

grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=3,                # 3-fold cross-validation
    n_jobs=-1,           # use all cores
    scoring='neg_mean_squared_error',
    verbose=2
)

grid_search.fit(X_train, y_train)
```

```python
print("Best parameters found:", grid_search.best_params_)
best_model = grid_search.best_estimator_
```

```python
y_pred = best_model.predict(X_test)

print(f"MAE:  {mean_absolute_error(y_test, y_pred):.2f}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_test, y_pred)):.2f}")
print(f"R²:   {r2_score(y_test, y_pred):.2f}")
```

```python
# Cross Validation
from sklearn.model_selection import cross_val_score
model = grid_search.best_estimator_

# Perform 5-fold cross-validation (you can adjust cv folds)
cv_scores = cross_val_score(
    model,
    X_train,
    y_train,
    cv=5,
    scoring='neg_mean_squared_error',  # use neg MSE for regression
    n_jobs=-1
)

# Convert negative MSE to positive RMSE
rmse_scores = np.sqrt(-cv_scores)

print("Cross-validation RMSE scores:", rmse_scores)
print("Mean RMSE:", rmse_scores.mean())
print("Standard deviation of RMSE:", rmse_scores.std())
```
